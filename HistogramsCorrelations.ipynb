```python
import pandas as pd
# import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
a = pd.read_excel('EnglishFiltered.xlsx')
# for ID in a.columns[1:]:
(fig,ax) = plt.subplots(10,figsize=(6,35))
for step,col in enumerate(a.columns[1:11]):
    ax[step].hist(a[col].values[a[col].values>0],bins=100)
    ax[step].set_title(col)
# plt.show()
```

```python
a = pd.read_excel('EnglishFiltered.xlsx')
# for ID in a.columns[1:]:
(fig,ax) = plt.subplots(10,10,figsize=(80,80))
for step,col in enumerate(a.columns[1:11]):
    for step2,col2 in enumerate(a.columns[1:11]):
        ax[step,step2].scatter(a[col].values,a[col2].values)
        prams = np.polyfit(a[col].values,a[col].values,1)
        ax[step,step2].plot([0,max(a[col].values)],[prams[0],prams[0]+prams[1]*max(a[col].values)])
        ax[step,step2].set_title(str(col)+'vs'+str(col2))
# plt.show()
```
