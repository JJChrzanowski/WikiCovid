```python
# name=names[0]
    df = pd.read_excel('raw/'+name+'.xlsx')
    df = df.loc[df[df.columns[0]]>=datetime.datetime(2020,1,1)]
    a = pd.DataFrame(df[df.columns[1:]].sum(axis=0))
    print(name,a.sort_values(by=a.columns[0],ascending=False).index[:11])

    
global_daysum = []
for name in names:
    npdf = np.array(df[df.columns[1:]])
    npdf_daysum = np.sum(npdf,axis=-1)
    df_daynorm = pd.DataFrame(100*np.apply_along_axis(np.divide,0,npdf,npdf_daysum))
    df_daynorm.columns = df.columns[1:]
    df_daynorm.index = df[df.columns[0]]
    df_daynorm.to_excel('daynorm/'+name+'_daynorm.xlsx')
    global_daysum.append([name,npdf_daysum])
    
df_global_daysum = pd.DataFrame(np.vstack([i[1] for i in global_daysum]))
df_global_daysum = df_global_daysum.T
df_global_daysum.columns = [i[0] for i in global_daysum]
df_global_daysum.index = df[df.columns[0]]
df_global_daysum.to_excel('daynorm/global_daysum.xlsx')

plt.figure(figsize=(5,3),dpi=600)
for name in names:
    plt.plot(df_global_daysum.index,df_global_daysum[name].rolling(window=30).mean(),label=name)
plt.legend(bbox_to_anchor=(0,0,1.2,0.95))
plt.ylabel('Count')
plt.xlabel('Dates')
plt.savefig('daysums.png')

df_global_daysum_ = df_global_daysum.loc[df_global_daysum.index>datetime.datetime(2019,9,1)]
plt.figure(figsize=(5,3),dpi=600)
for name in names:
    plt.plot(df_global_daysum_.index,df_global_daysum_[name].rolling(window=30).mean(),label=name)
plt.legend(bbox_to_anchor=(0,0,1.2,0.95))
plt.ylabel('Count')
plt.xlabel('Dates')
plt.savefig('daysums_september2019.png')

# plt.figure(figsize=(5,3),dpi=600)
# for name in names:
#     plt.plot(df_global_daysum.index,df_global_daysum[name].rolling(window=30).mean(),label=name)
# plt.legend(bbox_to_anchor=(0,0,1.2,0.95))
# plt.ylabel('Count')
# plt.xlabel('Dates')
# plt.savefig('daysums.png')


plt.figure(figsize=(5,3),dpi=600)
for year in [2015,2016,2017,2018,2019]:
    df_global_daysum_ = df_global_daysum.loc[(df_global_daysum.index>=datetime.datetime(year,1,1))&(df_global_daysum.index<datetime.datetime(year+1,1,1))]
    plt.plot([datetime.datetime(2019,1,1)+datetime.timedelta(days=int(_)-10) for _ in np.arange(len(df_global_daysum_))],df_global_daysum_['en'].rolling(window=30).mean(),label=year)
plt.legend(bbox_to_anchor=(0,0,1.3,0.75))
plt.ylabel('Count')
plt.xlabel('Days since beginning of the year')
plt.savefig('daysums_byyear_en.png')

# df_global_daysum.to_excel('daynorm/global_daysum.xlsx')

# # plt.plot(df_daynorm.index,np.sum(npdf,axis=-1))
# # # df_daynorm.Coronavirus.sum()
# # # df_daynorm.loc[df_daynorm.index>datetime.datetime(2020,1,1)].Coronavirus.plot()
```
